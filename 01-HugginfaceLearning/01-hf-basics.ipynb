```python
import os
from huggingface_hub import HfApi, create_repo, CommitOperationAdd

api     = HfApi(token=os.getenv("HF_WRITE_TOKEN"))
TOKEN   = os.getenv("HF_WRITE_TOKEN")
REPO_ID = "akshayredekar45/first-sentiment-analyzer"

try:
    api.delete_repo(repo_id=REPO_ID, repo_type="space", token=TOKEN)
    print("Old space deleted")
except Exception:
    print("No existing space found, continuing...")

create_repo(
    repo_id=REPO_ID,
    repo_type="space",
    space_sdk="gradio",
    private=False,
    token=TOKEN,
    exist_ok=False,
)
print("New space created")

README = b"""\
---
title: Simple Sentiment Analyzer
colorFrom: blue
colorTo: purple
sdk: gradio
sdk_version: "5.29.0"
app_file: app.py
pinned: false
license: mit
---

# Simple Sentiment Analyzer
Analyzes text as Positive, Negative, or Neutral.
"""

REQUIREMENTS = b"""\
# no gradio here - sdk_version in README controls it
"""

APP = b"""\
import gradio as gr

def analyze_sentiment(text):
    if not text.strip():
        return "Please enter some text."

    text_lower = text.lower()
    pos_words = ["good", "great", "happy", "love", "awesome", "excellent", "wonderful"]
    neg_words = ["bad", "hate", "sad", "terrible", "awful", "horrible", "worst"]

    p = sum(1 for w in pos_words if w in text_lower)
    n = sum(1 for w in neg_words if w in text_lower)

    if p > n:
        return f"Positive (score: +{p})"
    elif n > p:
        return f"Negative (score: -{n})"
    else:
        return "Neutral"

demo = gr.Interface(
    fn=analyze_sentiment,
    inputs=gr.Textbox(label="Enter text", placeholder="Type something..."),
    outputs=gr.Textbox(label="Result"),
    title="Simple Sentiment Analyzer",
    examples=[
        ["I love this, it is great!"],
        ["This is terrible and awful."],
        ["The weather is okay today."],
    ]
)

demo.launch()
"""

operations = [
    CommitOperationAdd(path_in_repo="README.md",        path_or_fileobj=README),
    CommitOperationAdd(path_in_repo="requirements.txt", path_or_fileobj=REQUIREMENTS),
    CommitOperationAdd(path_in_repo="app.py",           path_or_fileobj=APP),
]

api.create_commit(
    repo_id=REPO_ID,
    repo_type="space",
    commit_message="clean deploy: gradio 5.29.0",
    operations=operations,
    token=TOKEN,
)

print(f"Done. Live at: https://huggingface.co/spaces/{REPO_ID}")
```
