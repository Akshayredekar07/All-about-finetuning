```python
from transformers import pipeline

# ── 1. Load the pipeline ──────────────────────────────────────────────────────
pipe = pipeline(
    task="text-classification",
    model="distilbert-base-uncased-finetuned-sst-2-english",
    device=0
)

# ── 2. Single input ───────────────────────────────────────────────────────────
result = pipe("The movie was absolutely fantastic, I loved every moment of it.")
print("Single:", result)
# [{'label': 'POSITIVE', 'score': 0.9998}]
```

```python

# ── 3. List of inputs → automatic batch processing ────────────────────────────
results = pipe([
    "The food at this restaurant was delicious and the service was excellent.",
    "I regret buying this product. It broke after just two days of use.",
    "The weather today is neither too hot nor too cold, just average."
])
print("Batch:", results)
# [{'label': 'POSITIVE', 'score': 0.9997},
#  {'label': 'NEGATIVE', 'score': 0.9982},
#  {'label': 'POSITIVE', 'score': 0.6741}]

```

```python
# ── 4. batch_size for larger datasets ─────────────────────────────────────────
texts = [
    "I had an amazing experience at the hotel, staff was very helpful.",
    "The flight was delayed by 5 hours with no explanation given.",
    "This laptop is fast, lightweight, and the battery lasts all day.",
    "Customer support was rude and completely unhelpful.",
    "The book was a bit slow at first but the ending was truly surprising.",
    "Terrible packaging — the item arrived completely damaged.",
    "Best coffee I have ever had, I will definitely come back.",
    "The app crashes every time I try to open it on my phone.",
]

results = pipe(texts, batch_size=4)

# ── 5. Pretty-print all results ───────────────────────────────────────────────
for text, res in zip(texts, results):
    print(f"  [{res['label']}] ({res['score']:.4f})  →  {text}")
```

```python
from transformers import pipeline

# 1. distilbert-base-uncased-finetuned-sst-2-english
#    Binary sentiment: POSITIVE / NEGATIVE
#    Best for: product reviews, movie reviews, general opinions

pipe = pipeline("text-classification",
                model="distilbert-base-uncased-finetuned-sst-2-english")

results = pipe([
    "The movie was absolutely breathtaking, one of the best I have seen.",
    "The headphones broke after a week. Total waste of money.",
    "Delivery was late but the product itself is decent.",
])

for text, res in zip(results, results):
    print(f"[{res['label']}] ({res['score']:.4f})  →  {text}")

# [POSITIVE] (0.9998)  →  The movie was absolutely breathtaking...
# [NEGATIVE] (0.9995)  →  The headphones broke after a week...
# [NEGATIVE] (0.6821)  →  Delivery was late but the product itself is decent.

```

```python
# 2. cardiffnlp/twitter-roberta-base-sentiment
#    3-class sentiment: LABEL_0=Negative, LABEL_1=Neutral, LABEL_2=Positive
#    Best for: tweets, short social media posts

from transformers import pipeline
import torch

pipe = pipeline("text-classification",
                model="cardiffnlp/twitter-roberta-base-sentiment")

texts = [
    "Just got the new iPhone and I am OBSESSED 😍🙌",          # positive
    "Flight delayed again. Third time this month. So done.",    # negative
    "Watching the game tonight. Should be interesting.",        # neutral
]

results = pipe(texts, top_k=None)   # get all 3 scores

label_map = {"LABEL_0": "NEGATIVE", "LABEL_1": "NEUTRAL", "LABEL_2": "POSITIVE"}

for text, scores in zip(texts, results):
    top = max(scores, key=lambda x: x["score"])
    print(f"[{label_map[top['label']]}]  →  {text}")
    for s in scores:
        print(f"    {label_map[s['label']]}: {s['score']:.4f}")
```

```python

# 3. ProsusAI/finbert
#    Financial sentiment: positive / negative / neutral
#    Best for: earnings reports, financial news, analyst commentary

pipe = pipeline("text-classification",
                model="ProsusAI/finbert")

texts = [
    "The company reported record-breaking quarterly profits, exceeding all analyst expectations.",
    "Rising interest rates and supply chain disruptions led to a sharp decline in earnings.",
    "The board approved a share buyback program of up to 500 million dollars.",     # neutral
]

results = pipe(texts)

for text, res in zip(texts, results):
    print(f"[{res['label'].upper()}] ({res['score']:.4f})  →  {text}")

# [POSITIVE] (0.9921)  →  The company reported record-breaking quarterly profits...
# [NEGATIVE] (0.9887)  →  Rising interest rates and supply chain disruptions...
# [NEUTRAL]  (0.9134)  →  The board approved a share buyback program...
```

```python

# 4. martin-ha/toxic-comment-model
#    Toxicity detection: toxic / non-toxic
#    Best for: content moderation, comment filtering

from transformers import pipeline

pipe = pipeline("text-classification",
                model="martin-ha/toxic-comment-model")

texts = [
    "I completely disagree with your opinion but I respect your perspective.",  # non-toxic
    "You are a complete idiot and nobody wants to hear your stupid opinions.",   # toxic
    "This is the worst take I have ever read on this topic.",                    # borderline
]

results = pipe(texts, top_k=None)

for text, scores in zip(texts, results):
    top = max(scores, key=lambda x: x["score"])
    toxic_score = next(s["score"] for s in scores if s["label"] == "toxic")
    flag = "🚩" if toxic_score > 0.5 else "✅"
    print(f"{flag} [{top['label'].upper()}] toxic_score={toxic_score:.4f}  →  {text}")

# ✅ [NON-TOXIC] toxic_score=0.0023  →  I completely disagree with your opinion...
# 🚩 [TOXIC]     toxic_score=0.9978  →  You are a complete idiot...
# 🚩 [TOXIC]     toxic_score=0.7342  →  This is the worst take I have ever read...
```

```python

# 5. facebook/bart-large-mnli   (Zero-Shot Classification)
#    No fixed labels — you define the categories at runtime
#    Best for: topic classification, intent detection, custom categories

from transformers import pipeline

pipe = pipeline("zero-shot-classification",
                model="facebook/bart-large-mnli")

texts = [
    "The central bank raised interest rates by 50 basis points to combat inflation.",
    "Scientists discover a new species of deep-sea fish off the coast of New Zealand.",
    "The team secured a last-minute victory in the championship finals last night.",
]

candidate_labels = ["finance", "science", "sports", "politics", "technology"]

for text in texts:
    result = pipe(text, candidate_labels=candidate_labels)
    top_label = result["labels"][0]
    top_score = result["scores"][0]
    print(f"\n[{top_label.upper()}] ({top_score:.4f})  →  {text}")
    for label, score in zip(result["labels"], result["scores"]):
        print(f"    {label:12s}: {score:.4f}")
```
